# Scrapy eBay product scraper

This notebook runs a Scrapy spider that searches eBay for a query, scrapes a fixed number of result pages, and saves the products to a CSV file.


## 1. Set search options

Change `search_query`, `number_of_pages`, `output_file`, and `images_dir` before running the command cell.


```python
from pathlib import Path

search_query = "laptop"
number_of_pages = 2
output_file = "practice/ebay_laptop_products.csv"
images_dir = "practice/product_images"

script_path = "practice/ebay_products_spider.py"
if not Path(script_path).exists():
    script_path = "ebay_products_spider.py"
```

## 2. Run the spider

The spider code lives in `practice/ebay_products_spider.py`. Running Scrapy as a separate Python process makes the notebook easy to rerun.


```python
!python3 "{script_path}" --query "{search_query}" --pages {number_of_pages} --output "{output_file}" --images-dir "{images_dir}"
```
